```python
search = ['Cell','Hot-Spot','Cracking','Shadowing','Diode','Vegetation','Offline-Module','Soiling','No-Anomaly']
path = '/content/drive/MyDrive/Solar Module Diagnosis/InfraredSolarModules'
for j in search:
  for i in range(20000):
  #print(jsonObject[f'{i}']['anomaly_class'])
    if j in jsonObject[f'{i}']['anomaly_class']:
      img_name = jsonObject[f'{i}']['image_filepath']
      img = io.imread(f'{path}/{img_name}')
      img_cvt = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
      img_resize = cv2.resize(img, (200, 300))
      final_name = jsonObject[f'{i}']['image_filepath'].split('/')[1]
      cv2.imwrite(f'{path}/{j}/{final_name}', img_resize)
      cv2.waitKey(0)
      #print(jsonObject[f'{i}']['anomaly_class'])
```
